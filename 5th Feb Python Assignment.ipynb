Example :- smartphone on the basic of brand , model , storage and color. 

```python
class smartphone:
    def __init__(category , brand_name,model_number,storage,color):
        category.brand_name = brand_name
        category.model_number = model_number
        category.storage = storage 
        category.color = color
    

    def smartphone_details(category):
         return category.brand_name,category.model_number,category.storage,category.color
  
```

```python
phone1 = smartphone("Apple", "iPhone 14", "256GB", "Black")
```

```python
phone1.smartphone_details()
```
